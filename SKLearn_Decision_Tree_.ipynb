 memisahkan atribut dengan label. Selain itu, kita juga perlu
membagi dataset menjadi data latih dan data uji. Jalankan kode berikut untuk menerapkan
tahapan di atas.

```python
# Memisahkan atribut dan label
X = solar_system[['perihelion', 'aphelion', 'density', 'gravity', 'escape']]  # Ganti dengan kolom yang relevan
y = solar_system['eName']  # Nama satelit yang akan diprediksi


# Membagi dataset menjadi data latih & data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)

# Menampilkan ukuran data latih dan data uji
print("Ukuran data latih:", X_train.shape)
print("Ukuran data uji:", X_test.shape)
```

definisikan model decision tree yang akan kita gunakan. Kemudian,  latih model
menggunakan data latih menggunakan fungsi fit().

```python
from sklearn.tree import DecisionTreeClassifier

# Membuat model Decision Tree
tree_model = DecisionTreeClassifier()

# Melatih model dengan menggunakan data latih
tree_model.fit(X_train, y_train)  # Anda sudah benar di sini
```

```python
from sklearn.tree import DecisionTreeClassifier
# membuat model Decision Tree
tree_model = DecisionTreeClassifier()
# Melatih model dengan menggunakan data latih
tree_model = tree_model.fit(X_train, y_train)
```

```python
# Evaluasi Model
from sklearn.metrics import accuracy_score

# Melakukan prediksi pada data uji
y_pred = tree_model.predict(X_test)

# Menghitung akurasi
acc_score = round(accuracy_score(y_test, y_pred), 3)

# Menampilkan akurasi
print('Accuracy: ', acc_score)
```

```python
# Input untuk prediksi
input_data = [[1070400, 0.001, 0.177, 1.94, 1.428]]  # Pastikan ini sesuai dengan kolom yang dilatih

# Melakukan prediksi
prediksi = tree_model.predict(input_data)

# Menampilkan hasil prediksi
print("Hasil prediksi:", prediksi[0])
```
